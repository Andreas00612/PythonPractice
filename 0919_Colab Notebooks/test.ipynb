```python
def split_list(numlist, firstnum):
    first_list=[]
    second_list=[]
    for i in range(0,len(numlist)):
        if i<firstnum:
            first_list.append(numlist[i])
        else:
            second_list.append(numlist[i])
    return [first_list,second_list]

def third():
    numlist = [1, 6, 2, 4, 7, 10, 3]
    firstnum = 2
    print(split_list(numlist, firstnum))
third()
```
